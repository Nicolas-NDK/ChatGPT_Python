```python
#Para gerar o codigo abaixo foi realizada as seguintes perguntas:

#Escreva uma função em python que receba um arquivo .csv, crie um banco de dados de nome padrão com ajuda do sqlite, 
#crie uma tabela contendo os dados do dataframe mantendo os nomes das colunas, e salve o banco de dados num arquivo de saída.
#Como no codigo assima adicionamos uma janela para que a pessoa possa escolher o aquivo .CSV que ela deseja utilizando o seguinte comando.
#Funcinou perfeito, agora preciso que este codigo abra uma janela e a pessoa possa importar um arquivo csv para gerar o database

import tkinter as tk
from tkinter import filedialog, messagebox
from tkinter import ttk  # Importar ttk para utilizar o ttk.Button
import sqlite3
import pandas as pd

def csv_to_sqlite(csv_file, db_file):
    # Lê o arquivo CSV usando o pandas
    dataframe = pd.read_csv(csv_file)

    # Conecta-se ao banco de dados SQLite
    conn = sqlite3.connect(db_file)

    # Obtém o nome da tabela a partir do nome do arquivo CSV
    table_name = csv_file.split('.')[0]

    # Verifica se a tabela já existe no banco de dados
    cursor = conn.cursor()
    cursor.execute(f"SELECT name FROM sqlite_master WHERE type='table' AND name='{table_name}'")
    existing_table = cursor.fetchone()

    # Se a tabela existir, exclui-a do banco de dados
    if existing_table:
        cursor.execute(f"DROP TABLE IF EXISTS {table_name}")

    # Salva o dataframe no banco de dados
    dataframe.to_sql(table_name, conn, index=False)

    # Fecha a conexão com o banco de dados
    conn.close()

    print(f"Banco de dados {db_file} criado com sucesso!")

def import_csv():
    # Abre uma janela para selecionar o arquivo CSV
    csv_file = filedialog.askopenfilename(filetypes=[('CSV Files', '*.csv')])

    if csv_file:
        # Define o nome do arquivo de banco de dados SQLite de saída
        db_file = csv_file.split('.')[0] + ".db"

        # Chama a função para converter o CSV em banco de dados
        csv_to_sqlite(csv_file, db_file)

        # Exibe uma mensagem de sucesso
        messagebox.showinfo("Sucesso", f"Banco de dados {db_file} criado com sucesso!")

# Cria a janela principal
root = tk.Tk()

# Define a cor de fundo do botão
button_bg = "#3572CF"  # Vermelho

# Define a cor do texto do botão
button_fg = "#3572CF"  # Branco

# Define o estilo do botão
button_style = ttk.Style()
button_style.configure("CustomButton.TButton",
                       background=button_bg,
                       foreground=button_fg,
                       font=("Arial", 12, "bold"))

# Cria um botão para importar o arquivo CSV
button = ttk.Button(root, text="Importar CSV", command=import_csv, style="CustomButton.TButton")
button.pack(pady=10)

# Inicia o loop da interface gráfica
root.mainloop()
```
